```python
import matplotlib.pyplot as plt
import pandas as pd
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
picData = unrate.head(5)
```

```python
plt.plot(picData['DATE'],picData['VALUE'])
plt.xticks(rotation =45)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Month Unemployment Rate 1948')
plt.show()
```
